```python
# import libraries
 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Influenza Incidence in Germany in recent years

### Data Source
Robert Koch-Institut: SurvStat@RKI 2.0, https://survstat.rki.de, Retrieval date: 25. January, 2021.

SurvStat@RKI 2.0 is a web based platform to request data from the database of infectious diseases which are required to be registered in accordance to the German infectious diseases protection law, Infektionsschutzgesetz (IfSG).  

### Variables
The presented data shows the Influenza incidence per 100,000 people and per season week in Germany in the current and the four previous influenza seasons. The first season week starts at the 27th calendar week of the year and a season is defined as a year-long period beginning at the first season week.

For more information, please consult the `Influenza_Seasons_Info.pdf` file (German).

```python
df = pd.read_csv("Influenza_Seasons_Data.csv", encoding = "utf-16", sep='\t',skiprows = 1, decimal = ',')
df
```

```python
dfT = df.T
dfT = dfT.rename(columns=dfT.iloc[0])[1:].astype('float32').reset_index() 
```

### Data Visualization

Plot of the influenza seasons from 2016 to 2021. 

Note that due to social distancing measures and mandatory mask policies in Germany as well as lockdowns and restrictions for private and public meetings, the current 2020/2021 season has not started yet and could be strongly suppressed due to these temporary policies.

```python
ax = plt.gca() 

dfT.plot(kind='line',x='index',y='2016/17',ax=ax)
dfT.plot(kind='line',x='index',y='2017/18', ax=ax)
dfT.plot(kind='line',x='index',y='2018/19', ax=ax)
dfT.plot(kind='line',x='index',y='2019/20', ax=ax) 
dfT.plot(kind='line',x='index',y='2020/21', ax=ax) 

plt.style.use('seaborn')
plt.title('Influenza Incidence in Germany From 2016 to Jan. 2021, Source: RKI') 
plt.xlabel('Season Week')
plt.ylabel('Incidence per 100.000 people')

plt.show()
```
